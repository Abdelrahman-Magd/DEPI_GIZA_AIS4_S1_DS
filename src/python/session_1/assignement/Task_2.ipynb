```python
#Task_2 Encoded Message
message_1 = ("###!!@mocleW EPGTQ!!!6789") 
core_message = message_1[6:-7:1]
print(core_message)        # 1. Extract the core part of the message: "mocleW EPGTQ".




words = core_message.split()
print(words)
reversed_word_1 = words[0][::-1]
print(reversed_word_1)
list_char = list(reversed_word_1)
print(list_char)
list_char.append("e")
print(list_char)
fixed_word = "".join(list_char)
print(fixed_word)         # 2. Reverse the first word: "mocleW" becomes "Welcome".




second_word = words[1]
fix = second_word.removeprefix("E")
print(fix)                # 3. Remove shifted vowels in the second word

final_message_2 = fixed_word + " " + fix
print(final_message_2)    # 4. Final decoded message: "Welcome PGTQ"
```

```python
#Task_3 Encoded Message
message_3 = ("&&&**$gnirtS PLIO!!@1234")
core_message = message_3[6:-7:1]
print(core_message)      # 1. Extract the core part of the message: "gnirtS PLIO".




words = core_message.split()
print(words)
reversed_word = words[0][::-1]
print(reversed_word)     # 2. Reverse the first word: "gnirtS" becomes "String".




second_word = words[1]
fix = second_word.replace("PLIO", "PLEU")
print(fix)               # 3. Replace shifted vowels in the second word:




final_message = reversed_word + " " + fix
print(final_message)     # 4. Final decoded message: "String PLEU".
```

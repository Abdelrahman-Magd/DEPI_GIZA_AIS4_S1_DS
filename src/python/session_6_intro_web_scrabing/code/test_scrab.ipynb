```python
import csv
from time import sleep
from random import uniform
base_url = 'https://wuzzuf.net/search/jobs/?q=python&a=navbl'
query = "machine learning"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
numbr_of_pages = 5
with open('wuzzuf_jobs.csv', mode='w', newline='', encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["job Title","Company","location","posted","Skills/qualifications"])
    for page in range(numbr_of_pages):
        params = {
            'q': query,
            'start': page * 15
        }
        print(f"fteching page {page + 1}.......")
        response = requests.get(base_url, headers=headers, params = params)
        soup = BeautifulSoup(response.content, "html.parser")
        jobs = soup.find_all('div',{"class":"css-ghe2tq e1v1l3u10"})
        for job in jobs:
            title_tag = job.find('h2', class_ = 'css-193uk2c').text.strip()
            company_tag = job.find('a', class_ = 'css-o171kl').text.strip()
            location_tag = job.find('div', class_= 'css-1k5ee52').text.strip()
            posted_tag = job.find('div',class_ = 'css-eg55jf').text.strip()
            skill_tag = job.find('div', class_ ='css-1rhj4yg').text.strip()
            writer.writerow([title_tag,company_tag,location_tag,posted_tag,skill_tag])
            


    
```

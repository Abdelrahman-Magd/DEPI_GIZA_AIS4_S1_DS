# Thanos project
    make a function to do the following :
    1- open the whole folder
    2- create number of file inside it
    3- check the len of files in that folder
    4- delete random half of this folder
    5- check the result
    

```python
import os
import random


def Thanos_project():
    folder_path = "D:/DEPI/DEPI_GIZA_AIS4_S1_DS/src/python/session_4/assignement/Thanos_project"               


    if not os.path.exists(folder_path):                   ####     1- open the whole folder
        os.makedirs(folder_path)
    else:
        print(f"{folder_path} already exists")


    for i in range (40):                                  ####    2- create number of file inside it
        with open(f"{folder_path}/thanos_{i}.txt","a+") as file:
            file.write("Welcome to DEPI.")
    print("files created successfully.")


    files = os.listdir(folder_path)
    count1 = len(files)
    print(f"files count before is :{count1}")            ####    3- check the len of files in that folder


    deleted_number = count1 // 2                         ####    4- delete random half of this folder
    random.shuffle(files)
    print(f"files is being deleted is {deleted_number}")
    for i in range(deleted_number):
        os.remove(f"{folder_path}/{files[i]}")


    remaning_files = os.listdir(folder_path)              ####   5- check the result
    print(f"files deleted successfully files after this is : {len(remaning_files)}")







Thanos_project()
    
```

# Problem solving # 1
    Complete the solution so that it reverses all of the words within the string passed in.

    Words are separated by exactly one space and there are no leading or trailing spaces.

    Example(Input --> Output):

    "The greatest victory is that which requires no battle" --> "battle no requires which that is victory greatest The"

```python
def reversed_words(s):
    words = s.split(" ")
    reversed_list = words[::-1]
    sentence = " ". join(reversed_list)

    return sentence



s = ("battle no requires which that is victory greatest The")
reversed_words(s)
```

Another solution


```python
def reversed_words(s):
    return " ".join(s.split(" ")[::-1])

s = ("battle no requires which that is victory greatest The")
reversed_words(s)
```

# Problem solving # 2
    Create a function which answers the question "Are you playing banjo?".
    If your name starts with the letter "R" or lower case "r", you are playing banjo!

    The function takes a name as its only argument, and returns one of the following strings:

    name + " plays banjo" 
    name + " does not play banjo"
    Names given are always valid strings.

```python
def are_you_playing_banjo(name):
    if name[0].lower() == 'r':
        return name + " plays banjo"
    else:
        return name + " does not play banjo"
    

"""
TEST

"""

names = ["Rahma", "rami", "Adam", "mohamed"]

for name in names:
    print(are_you_playing_banjo(name))
```

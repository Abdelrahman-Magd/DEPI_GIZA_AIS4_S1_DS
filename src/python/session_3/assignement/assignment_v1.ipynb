# find the factorial of a number 
## Example
(factorial of 5) = 5 * 4 * 3 * 2 * 1 = 120

```python
# Write your answer here
num = int(input("Enter a number :"))
factorial = 1 
if num < 0 :
    print("factorial does not exist.")
elif num == 0 :
    print("factorial is 1")
else:
    for i in range(1,num+1):
        factorial = factorial*i
    print("factorial of ",num,"is",factorial)
```

# Create a function that Calculates the Sum of Even Numbers in a List
## Example
numbers: [1, 2, 3, 4, 5, 6]

Output: 12

```python
# Write your answer here
def sum_even_numbers(numbers):
    return sum(num for num in numbers if num%2 ==0) 


list =[1, 2, 3, 4, 5, 6]
sum_even_number = sum_even_numbers(list)
print(f"sum of even numbers is : {sum_even_number}")
```

# Create a function that takes a list as input and return reversed list

## Example
lst: [1, 2, 3]

Output: [3, 2, 1]

```python
def reverse_list(lst):
    return lst[::-1]

lst = [1,2,3]
reversed = reverse_list(lst)
print(f"list is : {lst}")
print(f"reversed list is : {reversed}")
```

# create a function that calculates the median of a list of numbers (a median is the middle element in sorted list of numbers)

## Example
numbers: [5, 2, 7, 3, 1]
